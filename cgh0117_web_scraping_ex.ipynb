```python
import requests
from bs4 import BeautifulSoup
import json
import re

# Target URL
URL = "https://supreme.justia.com/cases/federal/us/year/2025.html"

# Send GET request with a user-agent header
headers = {"User-Agent": "Mozilla/5.0 (compatible; JustiaScraper/1.0)"}
response = requests.get(URL, headers=headers)
response.encoding = "utf-8"

soup = BeautifulSoup(response.text, "html.parser")

# Container for all cases
cases_data = []

# Each case entry appears in the .results.zebra container
case_entries = soup.select("div.results.zebra.has-negative-sides-30.-overflow-hidden > div")

for entry in case_entries:
    case_info = {}

    # Case title
    title_el = entry.select_one("a")
    case_info["Case Title"] = title_el.get_text(strip=True) if title_el else None

    # Case URL
    case_info["Case URL"] = "https://supreme.justia.com" + title_el["href"] if title_el and title_el.has_attr("href") else None

    # Docket number
    docket_el = entry.select_one("strong")
    case_info["Docket Number"] = docket_el.get_text(strip=True) if docket_el else None

    # Date
    date_el = entry.select_one("span.date")
    case_info["Date"] = date_el.get_text(strip=True) if date_el else None

    # Court name (for Justia SCOTUS opinions, usually fixed)
    case_info["Court Name"] = "Supreme Court of the United States"

    # Opinion summary
    summary_el = entry.select_one(".truncate")
    case_info["Justia Opinion Summary"] = summary_el.get_text(strip=True) if summary_el else None

    # Clean up unwanted symbols or whitespace
    for k, v in case_info.items():
        if isinstance(v, str):
            v = re.sub(r"\s+", " ", v)
            v = re.sub(r"[^\x00-\x7F]+", "", v)  # remove non-ASCII
            case_info[k] = v.strip()

    cases_data.append(case_info)

# Save to JSON
with open("justia_supreme_court_2025.json", "w", encoding="utf-8") as f:
    json.dump(cases_data, f, ensure_ascii=False, indent=2)

print(f"✅ Scraped {len(cases_data)} cases and saved to justia_supreme_court_2025.json")
with open("justia_supreme_court_2025.json", "r", encoding="utf-8") as f:
    data = json.load(f)
    print(json.dumps(data, indent=2, ensure_ascii=False))
```
