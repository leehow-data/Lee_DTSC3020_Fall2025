## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    palette="Set2"
)
plt.title("Scatter Plot of Bill Length vs. Bill Depth by Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()

```

**Your interpretation (2–3 sentences):**

Adelie penguins tend to have more bill depth for shorter bills. Chinstrap penguins tend to have a more even ratio of length to depth. Gentoo penguins tend to have longer bills with less depth.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

plt.figure(figsize=(8,6))
sns.histplot(penguins["bill_length_mm"], bins=30, kde=True, color='skyblue')
plt.title("Histogram of Bill Length (30 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# TODO: Histogram with 30 bins
plt.figure(figsize=(8,6))
sns.histplot(penguins["bill_length_mm"], bins=10, kde=True, color='skyblue')
plt.title("Histogram of Bill Length (10 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")
plt.show()
```

**Your comparison (1–2 sentences):**

With thirty bins, it is more clear to see the chart is bimodal. Bill Lenghts peak around 40, 45, and 50 mm

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species
plt.figure(figsize=(8,6))
sns.boxplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    palette="pastel"
)
plt.title("Boxplot of Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()
```

**Your summary (1–2 sentences):**

Adelie and Chinstrap penguins have an average body mass of about 3750g, although the Chinstrap penguins have a bit less diversity. Gentoo penguins have a body mass averaging about 5100g.
